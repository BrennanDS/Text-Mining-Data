# Natural Language Toolkit

```python
f = open("data/goldBug.txt", "r")
goldBugString = f.read()
f.close()
print(goldBugString[:100])
```

This is supposed to show all the 'tokens' in the story:

```python
import nltk
goldBugTokens = nltk.word_tokenize(goldBugString.lower())
goldBugTokens[:10]
```

Do I want to treat words with different capitalizations as the same? If so, call the .lower function on the string above. 

Then, unless counting them for specific purposes, you should get rid of the punctuation marks.

```python
filterTokens = []
for word in goldBugTokens:
    if word.isalpha(): ## checks for alphabetic characters; but this got rid of GOLD-BUG too, because of the hyphen. 
        filterTokens.append(word) 
print(filterTokens)

print([word for word in goldBugTokens[:10]])
print([len(word) for word in goldBugTokens[:10]])
```

You can treat the list as a loop itself in this example:

This says, if a character in a word/token is alphabetic, add that word to 'goldBugWords'. 

```python
goldBugWords = [word for word in goldBugTokens if any([char for char in word if char.isalpha()])]
```

Asking for a frequency distribution, and then sorting it in descending order of frequency!

```python
wordFrequencies = nltk.FreqDist(goldBugWords)
wordFrequencies.most_common(10)
```

Now we'll call up a stopword list to parse words with low semantic but high syntactical value:

```python
stopwords = nltk.corpus.stopwords.words("English")
```

```python
print(stopwords)
```

```python
goldBugFilteredWords = [word for word in goldBugWords if not word in stopwords]
nltk.FreqDist(goldBugFilteredWords).most_common(10)
```

### NEXT STEPS

__Show a table of the top 20 words:__

```python
goldBugWordFreq = nltk.FreqDist(goldBugFilteredWords)
goldBugWordFreq.tabulate(20)
```

__Chose 3 words to add to the stopwords list using list concatenation.__ I'm chosing the first 3 because they have low semantic but high syntactic value, and for the purpose of this excercise I do not need them. 

```python
newStopWords = ["upon", "de", "'s"]
```

Now I'll concatenate the two lists together; I'll keep the list name 'stopwords' because I'll no longer need the old list.

```python
stopwords += newStopWords
```

__Now regenerate the above list:__

```python
NEWgoldBugFilteredWords = [word for word in goldBugFilteredWords if not word in stopwords]
goldBugWordFreq = nltk.FreqDist(NEWgoldBugFilteredWords)
goldBugWordFreq.tabulate(20)
```

__Now, instead of testing for presence in the stopword list, how would test for words that contain 10 characters or more?__ (Note: I'm not sure if this means we should find words that are longer of shorter than 10 characters -- I'm assuming that 'test for' here means to omit those that meet the subsequent requirement, i.e 10 chars.)

```python
shortWords = [word for word in goldBugWords if len(word) < 10]
print(shortWords[:20])
```
