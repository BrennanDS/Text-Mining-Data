## Why *Piers Plowman*?

William Langland's *Piers Plowman* is a late 14th-Century, allegorical, alliterative verse Dream Vision embodying the author's concerns with economic, religious, political, and judicial changes. The most cohesive 'narrative' traces the Dreamer's various (and sometimes multiply nested) visions as he undergoes a search for 'Dowel' and the elusive 'Truth', though various Passus break off to offer Biblical metaphors and retellings. Sometimes described as a medieval Ulysses, it is largely regarded as one of most complex Middle English texts. One of my professors told me that it's a poem scholars don't usually write books about, but individual articles; and when they do they tend to focus on select Passus rather than the whole (which is not to say that the whole is neglected or forgotten). The Passus and their according themes are often only tangentially related and the plot progresses in an extremely dream-like, sometimes fractured, manner. For this reason it is rather difficult to analyze the whole text . I think [this](http://virgil.org/dswo/courses/medieval-visions/pp-outline.htm) outline is one of the most complete and simple ones to help navigate the poem, and it provides good descriptions of what happens in each individual Passus.

With that in mind, I chose to use *Piers Plowman* mostly because I recently read it for a course on Medieval Law and Documentary culture and had many questions that I thought text mining tools could help answer. In terms of my familiarity with the text, I have read the whole thing once and several Passus multiple times for a paper on its allegorical tendencies. My past research, then, has focused more on the poetics and creative workings of the the poem than its polemics. While I think text mining could be extremely useful in clarifying or discovering some of Langland's alliterative tendencies (I'll talk more about that in the [next steps](#next_steps) section) I also think the convulutued Passus style narrative can perhaps be better understood with the techniques we've looked at so far. Specifically, I'm hoping to be able to compare the Passus to see in what ways they are or are not connected, and how several themes and keywords are dispersed throughout the poem.

### <a id='questions'></a> Interpretive Questions

One of the initial questions I had when chosing to look at *Piers Plowman* was whether there would be a noticeable diference in usage and placement of the words [*truthe*](http://quod.lib.umich.edu/cgi/m/mec/med-idx?type=id&id=MED47016) (just glimpsing at truthe's MED entry reveals its wide semantic range) and [*sothe*](http://quod.lib.umich.edu/cgi/m/mec/med-idx?type=id&id=MED41701). Richard Firth Green, in his fascinating book [*The Crisis of Truth*](http://www.upenn.edu/pennpress/book/746.html), examines the shifting meaning of *truthe* in the middle ages, and it's move away from the spoken human word to a heavy reliance on documentary proof. In the second chapter, he examines how *truthe* and *sothe* were once nearly interchangeable, but *truthe* eventually grew to refer to an objective, empirical thing; the Real. Before reaching this completely objective sense of the word that we might find in the Early Modern Period, it became highly associated with religious truth, with God's ultimate reality, existence, and providence. On the other hand, *sothe* was somewhat left behind, and maintained connotations of righteoussness, honour, and justice -- qualities to be exemplified through human action. Knowing the *sothe* of a testamony or of someone's word, for example, is more a question of how the individual's honour and reknown are perceived than how their statement corresponds to the perceptable world. When [looking into truthe](#truth), I'm hoping to draw out distinctions between the two words by using NLTK's collocation and concordance functions, as well as comparing the presence of truth related words across the Passus. I then hope to compare other meaningful lexical groups to see which tokens or groups most directly correlate to the presence of truth related words from Passus to Passus. 


At the very outset of this project, my main goal was to compare various versions of *Piers Plowman*. The 60+ original manuscripts that exist are separated into three distinct versions of the poem (A, B, and C texts; there's also hypothesized Z-version that all three might draw from but which has no known material existence). The creations of the B and C texts are belived to be about 15 years apart, with many significant socio-political changes (liek the 1381 Peasants' Revolt) occuring in between them, and I wanted to examine the B and C texts for differences in the way they treat some of the more political and religious ideas.  Unfortunately, due to the critical prominence of the B text, I couldn't find any electronic C versions. To compensate for that, my main focus is going to be on segementing the text by Passus, and trying to find connections and disparities between them; essentially, I want to see if there are any trends that can be traced with word distribution that would otherwise be difficult to detect when close reading such a muddled text.

### Tokenizing and Lemmatizing the text

I'l begin by tokenizing what I have in order to compare type/token ratios between Passus, do some frequency comparisons, and perhaps find some most common words after creating/editing a pseudo Middle English stopwords list. I tried finding an existing stopwords list for ME, but apparently none such things exist. After looking through ME DH tools, I found [this](http://scottkleinman.net/blog/2013/01/20/1568/) blog. The author, Scott Kleinman, has several posts pertaining to ME text mining and data analysis, all of which are really interesting, though he doesn't offer a stopword list for the taking. Thankfully, because I'm only looking at a single text for this project, scribal proclivities--especially when it comes to dialectical and regional differences--won't come into play much. Also, Langland wrote in the later medieval period and his language loosely resembles Chaucer's, so some of the spelling variations overlap with modern usage. I'll thus take nltk's normal english one and add a few (as many as I can think of) words to it. 

````python
import nltk

allTokensFromPassus = []
for passus in piersAllPassus:
    allTokensFromPassus += [word for word in nltk.word_tokenize(passus.lower()) if any([char for char in word if char.isalpha()])]
    
allTokensFromText = [word for word in nltk.word_tokenize(piersFullTextOnly.lower()) if any([char for char in word if char.isalpha()])]
````

These are two lists of tokens which I tried taking in two different ways because I am somehow losing certain instances of some types. I show this later with 'trouthe/truthe' in comparison to ctrl+f searching my original source. By using the 'allTokensFromText' I was able to get about 20 more occurences of the top 20 frequency words. The oder of them didn't change, though I still don't know what's causing this loss.

````python
stopwords = nltk.corpus.stopwords.words("English")
stopsToAdd = ["hym", "hem", "hyr", "hir", "hise", "hire", "alle", "thei", "thi", "thee", "ye", "thanne", "ben", "hadde", "hath", 
              "wolde", "us", "yow", "thow", "oure", "noght", "ne", "tho", "sholde", "ther", "thorugh", "whan", "al", "shal", 
              "ac", "er", "est", "quod", "seide", "both", "youre", "moore", "til", "thise", "thus", "may", "et", "many"] 
MEstopWords = stopwords + stopsToAdd

allTokensFromTextFiltered = [w for w in allTokensFromText if w not in MEstopWords]
top40filteredTokens = nltk.FreqDist(allTokensFromTextFiltered).most_common(40)
print(top40filteredTokens)
````

What I essentially did there was update the ```stopsToAdd``` list over and over again until a lot of the more common conjunctions and uneeded verbs were removed. However, sometmies I would add a word, like 'al', yet it wouldn't disappear from the most_common list. What I particularly enjoyed was the presence of the words "quod" and "seid" as the most frequent words, upholding the common notion that even in ME Dream Vision poetry those words can easily indicate a fictitious narrative. As is, I think the top 40 words listed above give a really good representation of the text, despite alternative spellings and homophones obscurring the 'real' count.

### <a id='comparing_collocations'></a>  Comparing Word Frequencies and Collocations

#### <a id='type_token'></a> Looking into type/token ratios per Passus

The first thing I thought I'd look at is a simple type/token ratio by passus on the offchance that something stands out and could possibly lead to new questions/paths for proceeding. 

````python
AmountOfTokensInEachPassus = [] ## while I didn't use all these variables, I thought I might as well include them here in case. 
AmountOfTypesInEachPassus = [] ## As is, this is only the number of types from the first 1500 tokens of each Passus.
AllTokensForEachPassus = []
TypeTokenRatioPerPassus = []

for passus in piersAllPassus:
    PassusTokens = [word for word in nltk.word_tokenize(passus.lower()) if any([char for char in word if char.isalpha()])]
    AllTokensForEachPassus.append(PassusTokens)
    AmountOfTokensInEachPassus.append(len(PassusTokens))
    PassusTypes = set(PassusTokens[0:1000]) ## I decided to change this to 1500 to account for variable Passus length
    AmountOfTypesInEachPassus.append(len(PassusTypes))
    TTR = int(round(len(PassusTypes)/10)) ## This seemed like the most efficient way of getting rounded integers.
    TypeTokenRatioPerPassus.append(TTR)   

print(AmountOfTokensInEachPassus, "\n", AmountOfTypesInEachPassus, "\n", TypeTokenRatioPerPassus)
````

````python
import matplotlib.pyplot as plt
%matplotlib inline

plt.bar(range(21), TypeTokenRatioPerPassus,color="#805180") # I later wrote a function to do this with better Xaxis labels, 
plt.ylabel("TTR %")                                         # however I needed some variables that I only define in the Truth section and didn't 
plt.xlabel("Passus Number")                                 # want to move this cell around as it would interupt the flow of the other section
plt.title("Type Tokens Ratios for each Passus")
plt.show()
````

While it may be a bit crude, and probably even non-informative, but I wanted to compare a simple type/token ratio per Passus to see if somthing striking could have been there. On my first attempt (before I limited the sample size to 1000 token) there were some drastic differences that I thought could have been telling of lexical density, but I quickly noticed they had a direct correlation to Passus length. Briefly reading more about Type Token Ratios showed that while there are more complicated and standardized equations to use, they usually aren't all that helpful. That was a bit disheartening after my initial reaction to the graph, but proved to be a useful learning experience in the end.

I'm not quite sure what words I expected to show up, but seeing 'dowel' (the imperative to 'do good work'; a significant aspect in Will the narrator's search for Truth), 'charite', 'love', and 'reson' appear next to 'truthe' is highly encouraging. What immediately stood out to me was that *truthe* (which is sometimes capitalized in *Piers Plowman* as a personified figured) was more closely aligned with spiritual words, while *sothe* was closer to more earthly ones. Whereas *truthe* is close to 'synne','soule, 'god', and 'christ', *sothe* is similar to 'peple/preest/kyng/plowman', 'sone/sonne' and 'commune/comune'. Langland's use of 'commune' to refer to both peasantry and the common court is another interesting lexical use examined by many scholars, and belonging to *sothe* can show *sothe* as having more earthly uses. This is further emphasized by the relation to the word 'pecok', which I think is certainly the strangest and my personal favourite word to appear here. It's used only a handful of times (5 according to my count, though 'pecokks' appears twice) and serves as a symbol for a rich man who neglects spiritual depth in favour of material values. In this sense, it might further tie *sothe* to wordly concerns which, if mistreated, could obscure more *truthely* ones. 

That being said, maybe a short concordance could help illustrate the differences further:

````python
PPTextFromTokens.concordance("truthe", lines = 5)
PPTextFromTokens.concordance("sothe", lines = 5)
````

From this small concordance we can see a similar disparity between the two words. *Truthe* is represented as something "therinne", as something to serve, and as "the beste". *Sothe*, from these examples, doesn't seem to convey direct notions of righteoussness or justice, but instead is often seen as something 'to know'. The commune 'knowen it wel, clerks wite it, and perhaps most strikingly, so does truthe. In this regard, I might argue that *sothe* is subordinate to truthe, suggesting that *truthe* holds more of a divine place while *sothe* serves a more immediate and human one. Indeed, this would fit well with the line we see in the concordance that says "whan alle tresors arn tried, truthe is the best".

While the two words have shown to have differnt contexts, I believe they can still both be used to find occurences of refernces to 'truth'. The distinctions are not to be forgotten, but I think grouping them together is justified since they each play important parts throughout the narrative. Specifically, after seeing the above examples I'm inclined to believe that *truthe* represents some ineffable, sacred value, and that *sothe* acts as an accessible bridge between it and the bisshops, plowmen, and clerks of the world.

Alternatively, I'd like to try to see how many occurences of these truthWords happen per Passus, and whether a trend can be established. To do this, I want to have an ordered dictionary with Passus# : Count entries. I'll first declare the variables I'll need, then make a loop to count the number of TruthWords per passus:

````python
y = 0
truthInPassusDict =  OrderedDict()
Totalcount = 0

for passus in piersAllPassus:
    PassusNumber = "Passus" + str(y)
    if PassusNumber == "Passus0":
        PassusNumber = "Prologue"
    PassusTokens = nltk.word_tokenize(passus.lower())
    TruthCountInPassus = 0
    for word in ListOfTruthWords:
        TruthCountInPassus += PassusTokens.count(word)
    y +=1
    Totalcount += TruthCountInPassus
    truthInPassusDict[PassusNumber] = TruthCountInPassus

Totalcount
````

````python
import numpy as np
````

````python
top40TokensOnly = [word[0] for word in top40filteredTokens if word[0] not in ListOfTruthWords] ## I don't want to compare truthe to the broader truthwords
````

````python
top40Tokens_CountPerPassus = OrderedDict()
for word in top40TokensOnly: ## this variable was used early on when testing my code to get all the lines from the HTML text
    allPassusWordCount = []
    for passus in piersAllPassus:
        tempWordCount = passus.count(word)
        allPassusWordCount.append(tempWordCount)
    top40Tokens_CountPerPassus[word] = allPassusWordCount
````

I decided to first get a dictionary with ```[the top 40 words: their frequency in each passus]```. Next, I'll use numpy's correlation coefficient function to see which ones are the most common, and the least. The following loop should but their ratios into a dictinoary, and then I'll graph them on a line chart to see which ones are high, average, and negative. 

````python
TrutheCorrelationsDict = OrderedDict()
correlationWords = [] # I thought I would need these two variables for graphing, but I didn't in the end.
correlationNumbers = [] #
for word, count in top40Tokens_CountPerPassus.items():
    cor = np.corrcoef(piesrLowerCaseWordTokenTruthCount, count)[0,1]
    correlationWords.append(word)
    correlationNumbers.append(cor)
    TrutheCorrelationsDict[word] = cor
````

````python
plt.title("Correlation Coefficients between Truthwords \nand the 40 Most Common Words")
nltk.FreqDist(TrutheCorrelationsDict).plot()
````

What's important to remember for this graph is that, unlike the NLTK.similar wordlists, this correlation is determined by the presenece of each word per Passus in relation to the earlier graph/count of truth related words per Passus. Seeing "lawe", "crist", and "god" at the very bottom (and in the negatives!) of the graph, while really surprising at first, does not then directly suggest that those words are not related to Truthe, and that our earlier .similar list (which contained those words) is problematic. Instead, that those words appear from the .similar function and as negative correlated coefficients might indicate that those words are 'paradigmatically' related to Truthe. That is to say that when one appears, the other is less likely to appear due to them being somewhat interchangeable; the semantic and syntactic context of their appearances may be similar, but the progression of their distribution across the Passus are not. 

And this is only looking at the negative coefficients. When looking at the top one, "Piers", Truthe's role as an astoudingly prominent and key aspect of the text is further solidified through its close association and paralleled occurences with the salvific protagonist.
