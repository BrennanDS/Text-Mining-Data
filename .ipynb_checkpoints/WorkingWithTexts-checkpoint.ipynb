# Working with Texts

Let's try to get the works of Poe here!

we use urllib.request to import data from another URL

.read = gets the data of the requested URL in a byte sequece
.decode = then decodes this sequence into 

```python
import urllib.request
poeUrl = "http://www.gutenberg.org/cache/epub/2147/pg2147.txt"
poeString = urllib.request.urlopen(poeUrl).read().decode()
```

Let's see what this looks like:

```python
print(poeString)
```

If I only want to see part of the text, I index the string by doing:

```python
print(poeString[1:73])
```

```python
poeString.find("with")
```

```python
poeString[136]
```

If I want to search from the END of the text, format it like so:

```python
print(poeString[-75:])
```

But the whole text file is huge! How do I find and analyze ONLY Poe's short story!?

```python
start = poeString.find("THE GOLD-BUG")
end = poeString.find("FOUR BEASTS IN ONE")
goldBugString = poeString[start:end]
# show start and end of goldBugString
print(goldBugString[:75] + " [rest of the text…] " + goldBugString[-75:])
```

```python
import os
directory = "data"
if not os.path.exists(directory):
    os.makedirs(directory)
```

```python
f = open("data/goldBug.txt", "w")
f.write(goldBugString)
f.close()
```
