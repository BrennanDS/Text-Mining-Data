```python
import urllib.request
aliceURL = "http://www.gutenberg.org/files/11/11-h/11-h.htm"
aliceString = urllib.request.urlopen(aliceURL).read()
```

We're going to make the alice HTML file a soup object which will help with the formatting and make it highly searchable and convenient. You need to download this first, as we did in class, through anaconda's CMD

```python
from bs4 import BeautifulSoup
aliceSoup = BeautifulSoup(aliceString)
aliceSoup
```

CSS markup follows a formula like \< tag class=""\> \~~~ text ~~~~ </closethetag. With this knowledge you can directly call the 'children' elements with .get_text and other formats to get specific parts of the text. 

aliceSoup.text gives you a bunch of styling and metadate information you don't really need. you don't want the head content, so just ask for the body:

```python
print(aliceSoup.body.text)
```

### Working with XML now!

First we'll get a file that we know is stored locally to work with:

```python
import nltk
rjFile = nltk.data.find("corpora/shakespeare/r_and_j.xml")
```

```python
rjFile
```

```python
from xml.etree.ElementTree import ElementTree
rjXml = ElementTree().parse(rjFile)
rjXml
```

Element tree can parse the contents of a file if it's XML. SO here we have an object, xml element tree, which we then parse the elements of the file into. 

```python
speeches = rjXml.findall(".//SPEECH")
len(speeches)
```

/ = root node

.// = start in the current node and anywhere in the document tree look for all the 'speech' nodes

```python
speakers = rjXml.findall(".//SPEAKER")
len(speakers)
```

```python
speakerNames = [speaker.text for speaker in speakers]
print(set(speakerNames))
```

finall returns the information with all those tags, you can use '.text' to get the text information in each tag!

```python
%matplotlib inline
nltk.FreqDist(speakerNames).plot(20)
```

```python
uniqueSpeakerNames = (set(speakerNames))
```

```python
titleCaseSpeakerNames = [word for word in uniqueSpeakerNames if word != None and word != word.upper()]
titleCaseSpeakerNames
```

```python
nltk.Text(speakerNames).dispersion_plot(titleCaseSpeakerNames)
```

```python
speakerDict = nltk.defaultdict(str)
speeches = rjXml.findall(".//SPEECH")
for speech in speeches:
    speaker = speech.find("SPEAKER").text
    for line in speech.findall("LINE"):
        if line.text:
            speakerDict[speaker]+=line.text+"\n"

            
print(speakerDict)
```

```python
speakersLengthDict = {}
for speaker, text in speakerDict.items():
    speakersLengthDict[speaker]=len(text)
    
speakersLengthDict
```

We have ths object called speakerDict which has all of our characters and all of their lines. So we can try to analyze some of the words between characters, and stuffs.

```python
RomeoTokens = nltk.word_tokenize(speakerDict["ROMEO"])
print(RomeoTokens.count("love"))
nltk.Text(nltk.word_tokenize(speakerDict["ROMEO"])).similar("love")

JulietTokens = nltk.word_tokenize(speakerDict["JULIET"])
print(JulietTokens.count("love"))
nltk.Text(nltk.word_tokenize(speakerDict["JULIET"])).similar("love")
```
