```python
# read Gold Bug plain text into string
f = open("data/goldBug.txt", "r")
goldBugString = f.read()
f.close()
```

```python
import nltk
goldBugTokens = nltk.word_tokenize(goldBugString)
len(goldBugTokens)
```

```python
goldBugTagged = nltk.pos_tag(goldBugTokens)
```

Let's try to generate a frequency graph of the parts of speech in the Gold Bug (how many verbs, nouns, etc., we'll use the existing tags as the are).

```python
%matplotlib inline
goldBugTags = [pos for word, pos in goldBugTagged]
goldBugTags = [pos if pos.isalpha() else "."  for pos in goldBugTags]
goldBugTagsFreqs = nltk.FreqDist(goldBugTags)
goldBugTagsFreqs.plot()
```

Let's create a dispersion plot of the parts of speech.

```python
goldBugTagsText = nltk.Text(goldBugTags)

goldBugTagsMostCommon = [tag for tag, freq in goldBugTagsFreqs.most_common(10)]
print(goldBugTagsMostCommon)
goldBugTagsText.dispersion_plot(goldBugTagsMostCommon)
```

```python
aliceString = nltk.corpus.gutenberg.raw("carroll-alice.txt")
aliceTokens = nltk.word_tokenize(aliceString)
```

How many adjectives per tokens are there in Alice in Wonderland?

```python
aliceTagged = nltk.pos_tag(aliceTokens)
```

```python
aliceTags = [pos for word, pos in aliceTagged]
aliceTags.count("JJ") / len(aliceTags)
```

```python
aliceNameTags = [word for word, pos in aliceTagged if "NNP" in pos]
nltk.FreqDist(aliceNameTags).most_common(20)
```

```python
aliceText = nltk.Text(aliceTokens)
aliceText.dispersion_plot(["Alice","Queen"])
```

```python
aliceWordPos = [word+"\t"+pos for word, pos in aliceTagged]
aliceWordPosFreqs = nltk.FreqDist(aliceWordPos)
aliceWordPosCounts = ["\t".join([term, str(count)]) for term, count in aliceWordPosFreqs.items() if count > 1]
for line in aliceWordPosCounts[:10]: # remove the index for all
    print(line)
```
